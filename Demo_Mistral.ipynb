Interactive & Context-Aware Prompting, where AI remembers past interactions, refines responses, evaluates confidence, and actively engages by asking clarifying questions.



```python
class ChatSession:
    """Maintains conversation history for context-aware AI responses."""
    def __init__(self, model="mistral"):
        self.model = model
        self.history = []

    def ask(self, prompt):
        """Sends a prompt to Ollama while maintaining context."""
        self.history.append({"role": "user", "content": prompt})
        response = ollama.chat(self.model, messages=self.history)
        self.history.append({"role": "assistant", "content": response["message"]["content"]})
        return response["message"]["content"]

# Initialize a new chat session
chat = ChatSession()
```

📝 Exercise 15: Memory & Context Retention

📌 Goal: AI remembers past discussions for a more natural conversation.

```python
print("🔹 Initial Question:")
print(chat.ask("What are the causes of climate change?"))

print("\n🔹 Follow-up Question:")
print(chat.ask("We talked about climate change earlier. Can you now suggest solutions?"))
```
